## Exercice 1: lists

Create a function `insert`, that produces the following results:

```
>>> l = [0, 9, 3, 10]
>>> print(insert(l, -1, 1))
[0, -1, 9, 3, 10]
>>> print(insert(l, -1, 2))
[0, 9, -1, 3, 10]
>>> print(insert(l, "epita", 4))
[0, 9, 3, 10, 'epita']
>>> print(insert(l, "epita", 10))
[0, 9, 3, 10, 'epita']
>>> print(l)
[0, 9, 3, 10]
```

````python
def insert(l,item,index):
    emptylist=[]
    if index>len(l):
        emptylist=l.copy()
        emptylist.append(item)
        return emptylist
    for i in range(len(l)+1):
        if i==index:
            emptylist.append(item)
        if i<len(l):
            emptylist.append(l[i])
    return emptylist

def insert_fast(l,item,index):
    return l[:index]+[item] + l[index:]

l = [0, 9, 3, 10]
print(insert_fast(l, -1, 1))

print(insert_fast(l, -1, 2))

print(insert_fast(l, "epita", 4))

print(insert_fast(l, "epita", 10))

print(l)

````
