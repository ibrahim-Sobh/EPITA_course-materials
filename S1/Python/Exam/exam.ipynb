````python
## WRITE IN THIS CELL YOUR CODE WITHOUT ANY ERRORS. YOU SHOULD GENERATE THE FUNCTION: inter_in
## question 2

def inter_in(l, i, j):
    if i > len(l) or j > len(l):
        return []
    first=l[i]
    second=l[j]
    if i < j :
        l[i]=second
        l[j]=first
    else:
        l[i]=second
        l[j]=first
        
inter_in(list_, 0, 10)
print(list_[0])
````

````python
## WRITE IN THIS CELL YOUR CODE WITHOUT ANY ERRORS. YOU SHOULD GENERATE THE FUNCTION: inter_out
## question 3 (BONUS)


def inter_out(l, i, j):
    if i > len(l) or j > len(l):
        return l
    first=l[i]
    second=l[j]
    if i < j :
        return l[:i] + [second] + l[i+1:j] + [first] +l[j+1:]
    else:
        return  l[:j] + [first] + l[j+1:i] + [second] +l[i+1:]

list_=inter_out(list_, 0, 10)
print(list_[0])
````
