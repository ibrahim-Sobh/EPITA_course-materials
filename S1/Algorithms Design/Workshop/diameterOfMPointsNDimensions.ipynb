### 1.Diameter computation of M points of N dimensions

```python
import numpy as np 

def longestPointFromP(Points,P):
    return np.argmax(np.sum(np.abs(Points-P),axis=1))


def getPerimeter(P):
    stop=False
    pickup=np.random.randint(len(P))
    #print(pickup)
    P1=P[pickup]
    #print(P1)
    P2=P[longestPointFromP(P,P1)]
    #print(P2)
    while stop==False:
        temp=P[longestPointFromP(P,P2)]
        if np.sum(np.abs(P1-temp),axis=0) != 0:
            P1=P2.copy()
            P2=temp.copy()
        else:
            stop=True
    return np.array([P1,P2])
#np.sum(np.abs(P1-P2),axis=0)

boundaries=100
#Test of M points for N dimensions
m=10**6
dimension=5

#example of n for 3 plot
#m=10**3
#dimension=3

np.random.seed(2)
P=np.random.randint(-boundaries,boundaries,dimension*m).reshape(m,dimension)
#print(P)
%time Extremities=getPerimeter(P)
P1=Extremities[0]
print('P1',P1)
P2=Extremities[1]
print('P2:',P2)
print('Diameter defined by the extremes P1 and P2:',np.sum(np.abs(P1-P2),axis=0))
```
