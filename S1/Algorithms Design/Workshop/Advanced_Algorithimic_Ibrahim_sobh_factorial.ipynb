```python
# This funcion is for calculating the factorial
def fact(number):
    if number<2:
        return number
    fact=1
    while number>1:
        fact= number*fact
        number=number-1
    return fact

# This function will calculate the number of leading zeros of a number
def count_zeros(n):
    countzeros=0
    if n==0:
        return 1;
    while n>0:
        reminder=n % 10
        if reminder != 0:
            return countzeros
        countzeros+=1
        n=n//10
    return countzeros

# Method Number 2
#There is a pattern between number of zeros and a fact of a certain number
# N ( number of zeros) = number of 2's and number of 5's in common =  2^a * 5 ^b * constant

def count_zeros_pattern(n):
    count5=0
    count2=0
    for i in range(1,n+1):
        temp= i
        while temp%5==0 and temp>1:
            count5+=1
            temp=temp//5 
        temp= i
        while temp%2==0 and temp>1:
            count2+=1
            temp=temp//2  
    return min(count2,count5);

# ----> Best one # the fastest one of them 
# Method Number 3
#There is a pattern between number of zeros and a fact of a certain number
# N ( number of zeros) = the sum of the division on base 5 of a factorial
# N / 5^1, 5^2 ..... unil we reach zero, the sum of the basis division will get us ower zero count

def count_zero_5_pattern(n):
    count=0
    multp=1
    temp=n
    while temp>1:
        temp=n
        temp=temp//(5**multp)
        count+=temp
        multp=multp+1
    return count


# BEST ONE

n=int(input())
for i in range(n):
    print(count_zero_5_pattern(int(input())))
```
