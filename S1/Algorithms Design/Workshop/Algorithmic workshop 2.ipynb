```python
def invert(n):
    rev=0
    while n!=0:
        remainder = n % 10
        rev *= 10
        rev+=remainder
        n =n//10
    return rev

million=11**50000
%time print(invert(int(million)))
```

```python
def invert1(num):
    digitlist = list(str(num))
    return int(''.join(digitlist[::-1]))

#million=3256**300000
million=3256**3
%time invert1(million)
```
