```python
import pandas as pd


pokemons = pd.read_csv('All_Pokemon.csv', sep=',', index_col="Number")

df = pd.DataFrame(pokemons, columns=["Name", "Generation", "Type 1", "Type 2", "Height", "Weight", "BMI"])
                  

types= df.groupby(by=["Type 1","Type 2"]).count()

types_1=(df.groupby(['Type 1','Type 2']).size() 
   .sort_values(ascending=False) 
   .reset_index(name='count') 
 )
types_1.sort_values(by=['Type 1','Type 2','count'],ascending=True)

df[df["Type 1"]==""]
```

```python
types_2= types_1.copy()
maxima = types_1.groupby('Type 1')['count'].max()
types_2['max'] = types_2['Type 1'].map(maxima)
Dominant_Types=types_2[types_2["count"]==types_2["max"]][["Type 1", "Type 2","max"]]
Dominant_Types.sort_values(by=['Type 1','Type 2','max'],ascending=True,inplace=True)
Dominant_Types
```

```python
maxima = types_1.groupby('Type 1')['count'].max()
dominant=types_1[types_1["count"]==types_1['Type 1'].map(maxima)]
dominant
```
