```
x = torch.tensor([1.1], requires_grad=True)
y = torch.tensor([0.5], requires_grad=True)
```

```
intermediate_tensor = x * y + x
```

```
print(intermediate_tensor.requires_grad)
```

```
loss = intermediate_tensor + torch.exp(y) * 2
```


$ \frac{\partial{loss}}{\partial{x}} $


$ \frac{\partial{loss}}{\partial{y}} $

```
loss.backward() # it computes the partial derivatirve of looss regarding to x and y 
```

```
t = torch.tensor([1.1, 0.1, -2])

relu_t = torch.tensor([1.1, 0.1, 0])


def relu(t):

    result = t.clone()
    result[result<0] = 0

    return result


relu(t), t
```

```
def relu2(t):

    zeros = torch.zeros_like(t)
    return torch.max(t, zeros)

relu2(t)
```
